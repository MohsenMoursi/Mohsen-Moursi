```python
import pandas as pd 
import warnings 
warnings.filterwarnings('ignore')
```

```python
# let's generate links to the other pages
def get_glasses_from_website(): 
    classes = pd.DataFrame()
    for num in range (1,6) : 
        website = 'https://ischool.syr.edu/classes/?page=' + str(num)
        data = pd.read_html(website)
        classes = classes.append (data[0])
    return classes 
#classes = get_glasses_from_website()  
#if I want to change the colums , classes.columns = ['sd','sdds','sdda'] in the same line with retur 
#classes.tail()
def engineer_grad_ungrad(classes):
    classes['coursenum'] = classes ['Course'] .str[3:]
    classes['Type'] = 'none'
    classes['Type'][classes['coursenum'].astype(int) < 500] = 'U'
    classes['Type'][classes['coursenum'].astype(int) >= 500] = 'G'
    return classes 

```
